```python
from sklearn.impute import KNNImputer
import numpy as np 

df.replace("?", np.nan, inplace=True)
df = df.apply(pd.to_numeric, errors='coerce')

imputer = KNNImputer(n_neighbors=2)
df_imputed = imputer.fit_transform(df)
df = pd.DataFrame(df_imputed, columns=df.columns)
df
```

outliers detection using boxplot

```python
# Outliers detection using boxplot
def remove_outliers_boxplot(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[column] > lower_bound) & (data[column] < upper_bound)]
    return data

# Assuming 'feature_to_remove_outliers' is the column you want to handle outliers for
df = remove_outliers_boxplot(df, 'T3')
df = remove_outliers_boxplot(df, 'T4U')
df = remove_outliers_boxplot(df, 'TT4')
df = remove_outliers_boxplot(df, 'FTI')
df = remove_outliers_boxplot(df, 'TSH')

# Outliers detection and removal using percentile methods
def remove_outliers_percentile(data, column, lower_percentile, upper_percentile):
    lower_bound = data[column].quantile(lower_percentile)
    upper_bound = data[column].quantile(upper_percentile)
    data = data[(data[column] > lower_bound) & (data[column] < upper_bound)]
    return data

# Assuming 'feature_to_remove_outliers' is the column you want to handle outliers for
df = remove_outliers_percentile(df, 'T3', 0.05, 0.95)
df = remove_outliers_percentile(df, 'T4U', 0.05, 0.95)
df = remove_outliers_percentile(df, 'TT4', 0.05, 0.95)
df = remove_outliers_percentile(df, 'TSH', 0.05, 0.95)
df = remove_outliers_percentile(df, 'FTI', 0.05, 0.95)
df
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

# Check the number of unique classes in y
num_classes = len(y.unique())
print("Number of unique classes in y:", num_classes)

# Create an XGBoost classifier for multi-class classification
xgb_classifier = xgb.XGBClassifier(
    objective='multi:softmax',  # Multi-class classification
    num_class=num_classes,      # Number of unique classes in y
    n_estimators=100,           # Number of boosting rounds
    random_state=42
)

# Train the model on the training set
xgb_classifier.fit(x_train, y_train)

# Make predictions on the testing set
y_pred = xgb_classifier.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_rep)
```
